```python
# Option 3 - Impute the missing values 
data_impute = data.assign(Embarked=data['Embarked'].fillna('Unknown'))
# data_impute = data['Age'].fillna(data['Age'].mean())
# data_impute = data['Age'].fillna(data['Age'].median())

data_impute
```

```python
data_impute['Age'] = data['Age'].fillna(data_impute.groupby('Pclass')['Age'].transform('median'))
```

```python
data_impute.isnull().sum()
```

```python
data_impute.drop(columns = ['Cabin'], inplace = True)
```

```python
data_impute.isna().sum()
```
